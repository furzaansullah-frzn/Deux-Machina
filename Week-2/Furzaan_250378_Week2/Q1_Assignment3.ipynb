```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\furza\\Deux-Machina\\Week-2\\Dataset\\extended_salary_data.csv")
```

```python
plt.scatter(df["YearsExperience"],df["Salary"])
plt.xlabel("Experience(in yrs)")
plt.ylabel("Salary")
```

```python
mean = np.mean(df["Salary"])
median = np.median(df["Salary"])
variance = np.var(df["Salary"])
sd = variance**0.5
print(f"Mean Salary = {mean}")
print(f"Median Salary = {median}")
print(f"Variance = {variance}")
print(f"Standard Deviation = {sd}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()
y = (df["Salary"]).to_numpy()
x = (df["YearsExperience"]).to_numpy().reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
model.fit(x_train, y_train)
```

```python
print(f"Slope = {model.coef_}")
print(f"Intercept = {model.intercept_}")
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_pred, y_test)
mse = mean_squared_error(y_pred,y_test)
print(f"Mean Absolute Error = {mae}")
print(f"Mean Squared Error = {mse}")
```
